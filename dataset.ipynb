```python
def product_counts(ProductID,dataframe:pd.DataFrame):
    return hlp.filter_product(ProductID,dataframe).shape[0]

def store_counts(StoreID,dataframe:pd.DataFrame):
    return hlp.filter_product(StoreID,dataframe).shape[0]
```

```python
forecast_product_counts = [product_counts(i,forecast_edit) for i in products_id['ProductID']]
past_forecast_product_counts = [product_counts(i,past_forecast_edit) for i in products_id['ProductID']]
previous_orders_product_counts = [product_counts(i,previous_orders_edit) for i in products_id['ProductID']]
sales_product_counts = [product_counts(i,sales_edit) for i in products_id['ProductID']]
store_stock_product_counts = [product_counts(i,store_stock_edit) for i in products_id['ProductID']]
```

```python
dc_intake_product_counts = [product_counts(i,dc_intake_edit) for i in products_id['ProductID']]
dc_stock_product_counts = [product_counts(i,dc_stock_edit) for i in products_id['ProductID']]
```

```python
product_occurance = pd.DataFrame({'ProductID':products_id['ProductID'],
                                  'forecast':forecast_product_counts,
                                  'past_forecast':past_forecast_product_counts,
                                  'previous_orders':previous_orders_product_counts,
                                  'sales':sales_product_counts,
                                  'store_stock':store_stock_product_counts,
                                  'dc_intake':dc_intake_product_counts,
                                  'dc_stock':dc_stock_product_counts})

product_occurance = product_occurance.sort_values(by=['forecast','past_forecast','previous_orders','sales','store_stock','dc_intake','dc_stock'], ascending=False)
product_occurance
```

```python
product_occurance.to_csv('data_edit/product_occurance.csv',index=False)
```
